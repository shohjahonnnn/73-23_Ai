```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

# Ma'lumotlar: maydon (m²) va narxi
X1 = np.array([[40], [50], [60], [70], [80], [90]])
y1 = np.array([15000, 20000, 30000, 40000, 50000, 60000])

model1 = DecisionTreeRegressor(max_depth=3)
model1.fit(X1, y1)

plt.figure(figsize=(10, 6))
plot_tree(model1, filled=True, feature_names=["Maydon (m²)"], rounded=True)
plt.title("Qaror Daraxti 1: Maydon va Narx")
plt.show()
```

```python
# Ma'lumotlar: xonalar soni va narxi
X2 = np.array([[1], [2], [3], [4], [5], [6]])
y2 = np.array([12000, 25000, 37000, 48000, 60000, 72000])

model2 = DecisionTreeRegressor(max_depth=3)
model2.fit(X2, y2)

plt.figure(figsize=(10, 6))
plot_tree(model2, filled=True, feature_names=["Xonalar soni"], rounded=True)
plt.title("Qaror Daraxti 2: Xonalar soni va Narx")
plt.show()
```
