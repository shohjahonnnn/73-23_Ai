```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Ma'lumotlar
data = {
    'xonalar_soni': [1, 2, 3, 4, 5, 6, 7],
    'narxi': [12000, 25000, 37000, 48000, 60000, 75000, 90000]
}
df = pd.DataFrame(data)

# 2. Jadvalni chiqarish
print("Ma'lumotlar to'plami:")
print(df)

# 3. DataFrame haqida umumiy ma'lumot
print("\nDataFrame haqida umumiy ma'lumot:")
print(df.info())

# 4. Statistik tahlil
print("\nStatistik ma'lumotlar:")
print(df.describe())

# 5. Vizualizatsiya (scatter plot)
plt.scatter(df['xonalar_soni'], df['narxi'], color='blue', label="Haqiqiy ma'lumotlar")
plt.xlabel("Xonalar soni")
plt.ylabel("Narxi")
plt.title("Xonalar soni va narxi o'rtasidagi bog'liqlik")
plt.grid(True)
plt.legend()
plt.show()

# 6. Model qurish
X = df[['xonalar_soni']]
y = df['narxi']
model = LinearRegression()
model.fit(X, y)

# 7. Chiziqli regressiya chizig'i
plt.scatter(X, y, color='blue', label="Haqiqiy ma'lumotlar")
plt.plot(X, model.predict(X), color='red', label="Bashorat chizig'i")
plt.xlabel("Xonalar soni")
plt.ylabel("Narxi")
plt.title("Chiziqli regressiya natijasi")
plt.grid(True)
plt.legend()
plt.show()
```
