```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Generate data
np.random.seed(42)
X = np.linspace(-10, 10, 100)
y_true = 2*X**2 + 3*X + 1

# Add some noise to create "real" data
noise = np.random.normal(0, 10, size=X.shape)
y_real = y_true + noise

# Split into train and test (for demonstration)
X_train, X_test = X[:80], X[80:]
y_train, y_test = y_real[:80], y_real[80:]

# Fit a polynomial model (degree 2)
coefficients = np.polyfit(X_train, y_train, 2)
poly_model = np.poly1d(coefficients)
y_pred = poly_model(X_test)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"RÂ² Score: {r2}")

# Create the plot with different style
plt.figure(figsize=(10, 6), facecolor='#f5f5f5')

# Plot real data with different marker style
plt.scatter(X_test, y_test, color='#2c7bb6', marker='D', s=50,
            edgecolor='black', linewidth=0.5, label='Real data')

# Plot model prediction with dashed line
plt.plot(X_test, y_pred, color='#d7191c', linestyle='--',
         linewidth=2.5, label='Model prediction')

# Plot the true parabola (without noise) for reference
plt.plot(X, y_true, color='#008837', alpha=0.7,
         label='True parabola: y=2xÂ²+3x+1')

# Customize the plot
plt.title("Parabola Model Performance", fontsize=14, pad=20)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)
plt.legend(frameon=True, facecolor='white', framealpha=0.9)

# Custom grid and spines
plt.grid(True, linestyle=':', alpha=0.7)
ax = plt.gca()
ax.set_facecolor('#f9f9f9')
for spine in ax.spines.values():
    spine.set_edgecolor('#cccccc')

plt.tight_layout()
plt.show()
```
